```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("data.csv")
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.info()
data.describe()
```

```python
data.plot(x = 'price_range' , y ='ram' , kind = 'scatter' , color = 'red')
plt.show()
```

```python
X = data.drop('price_range',axis=1)
Y = data['price_range']
Y.head()
```

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
X_std = std.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X_std, Y, test_size=0.3, random_state=1)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix

gnb = GaussianNB()
gnb.fit(X_train, Y_train)
Y_pred = gnb.predict(X_test)

gnb_accuracy = accuracy_score(Y_test, Y_pred)

print("Accuracy: ", accuracy_score(Y_test, Y_pred))
print(confusion_matrix(Y_test, Y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from matplotlib.pyplot import figure

dt = DecisionTreeClassifier()
dt.fit(X_train, Y_train)
Y_pred = dt.predict(X_test)

plt.subplots(figsize=(20, 15))  
plot_tree(dt.fit(X_train, Y_train), fontsize=7)
plt.show()

dt_accuracy = accuracy_score(Y_test, Y_pred)

print("Accuracy: ", accuracy_score(Y_test, Y_pred))
print(confusion_matrix(Y_test, Y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(X_train, Y_train)
Y_pred = (knn.predict(X_test))

knn_accuracy = accuracy_score(Y_test, Y_pred)

print("Accuracy: ", accuracy_score(Y_test, Y_pred))
print(confusion_matrix(Y_test, Y_pred))
```

```python
#knn non-standardized
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=1)

knn_non_standardized = KNeighborsClassifier()
knn_non_standardized.fit(X_train, Y_train)
Y_pred = (knn_non_standardized.predict(X_test))


print("Accuracy: ", accuracy_score(Y_test, Y_pred))
print(confusion_matrix(Y_test, Y_pred))
```

```python
import matplotlib.pyplot as plt

figure = plt.figure()
axis = figure.add_axes([0, 0, 1, 1])
classification_algorithms = ['naive bayes', 'tree', 'knn']
accuracy = [gnb_accuracy, dt_accuracy, knn_accuracy]
axis.bar(classification_algorithms, accuracy)
plt.show()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_std, Y, test_size=0.15, random_state=1)

gnb_low = KNeighborsClassifier()
gnb_low.fit(X_train, Y_train)
Y_pred = (gnb_low.predict(X_test))
gnb_low_accuracy = accuracy_score(Y_test, Y_pred)

dt_low = DecisionTreeClassifier()
dt_low.fit(X_train, Y_train)
Y_pred = dt_low.predict(X_test)
dt_low_accuracy = accuracy_score(Y_test, Y_pred)

knn_low = KNeighborsClassifier()
knn_low.fit(X_train, Y_train)
Y_pred = (knn_low.predict(X_test))
knn_low_accuracy = accuracy_score(Y_test, Y_pred)

X_train, X_test, Y_train, Y_test = train_test_split(X_std, Y, test_size=0.6, random_state=1)

gnb_high = KNeighborsClassifier()
gnb_high.fit(X_train, Y_train)
Y_pred = (gnb_high.predict(X_test))
gnb_high_accuracy = accuracy_score(Y_test, Y_pred)

dt_high = DecisionTreeClassifier()
dt_high.fit(X_train, Y_train)
Y_pred = dt_high.predict(X_test)
dt_high_accuracy = accuracy_score(Y_test, Y_pred)

knn_high = KNeighborsClassifier()
knn_high.fit(X_train, Y_train)
Y_pred = (knn_high.predict(X_test))
knn_high_accuracy = accuracy_score(Y_test, Y_pred)


print("0.15 test-size accuracy: ")
print("Naive Bayes: ", gnb_low_accuracy)
print("Decision tree: ", dt_low_accuracy)
print("Knn: ", knn_low_accuracy)
print("---------------")
print("0.3 test-size accuracy: ")
print("Naive Bayes: ", gnb_accuracy)
print("Decision tree: ", dt_accuracy)
print("Knn: ", knn_accuracy)
print("---------------")
print("0.5 test-size accuracy: ")
print("Naive Bayes: ", gnb_high_accuracy)
print("Decision tree: ", dt_high_accuracy)
print("Knn: ", knn_high_accuracy)
print("---------------")
```
